# Price

```python
bayot['Price'] = bayot['Price'].str.replace(',','',regex=False)
bayot['Price'] = bayot['Price'].str.replace('فیلا','',regex=False)
bayot['Price'] = bayot['Price'].str.replace('شقة','',regex=False)
```

```python
bayot['Price'] = bayot['Price'].astype('float')
```

# Rooms

```python
bayot['Rooms'] = bayot['Rooms'].str.replace('غرف','',regex=False)
bayot['Rooms'] = bayot['Rooms'].str.replace('غرفة','',regex=False)
bayot['Rooms'] = bayot['Rooms'].str.replace('ة','',regex=False)
bayot['Rooms'] = bayot['Rooms'].str.replace('استوديو', '1',regex=False)
bayot['Rooms'] = bayot['Rooms'].str.replace('حمام', '',regex=False)
bayot['Rooms'] = bayot['Rooms'].str.replace('حمامات', '',regex=False)
bayot['Rooms'] = bayot['Rooms'].str.strip()
```

```python
bayot = bayot[~bayot['Rooms'].str.contains('متر مربع', na=False)]
```

```python
bayot['Rooms'] = bayot['Rooms'].str.replace('ات', '', regex=False)
```

```python
bayot['Rooms'] = bayot['Rooms'].astype(str).str.strip().str.normalize('NFKC')
bayot = bayot[bayot['Rooms'] != 'بيوت - QjOM2g-202697308']
```

```python
bayot['Rooms'] = bayot['Rooms'].astype('float')
```

# Bathrooms

```python
bayot['Bathrooms'] = bayot['Bathrooms'].str.replace('94 متر مربع','2',regex=False)
bayot['Bathrooms'] = bayot['Bathrooms'].str.replace('54 متر مربع','2',regex=False)
```

```python
bayot['Bathrooms'] = bayot['Bathrooms'].str.replace('حمام', '',regex=False)
bayot['Bathrooms'] = bayot['Bathrooms'].str.replace('حمامات', '',regex=False)
bayot['Bathrooms'] = bayot['Bathrooms'].str.strip()
```

```python
bayot['Bathrooms'] = bayot['Bathrooms'].str.replace(r'.*متر مربع.*', '1', regex=True)
```

```python
bayot['Bathrooms'] = bayot['Bathrooms'].str.replace('ات', '',regex=False)
```

```python
bayot['Bathrooms'] = bayot['Bathrooms'].astype('float')
```

# Area

```python
bayot['Area'] = bayot['Area'].str.replace('متر مربع','',regex=False)
```

```python
bayot['Area'] = bayot['Area'].str.replace(',','',regex=False)
```

```python
bayot['Area'] = bayot['Area'].astype('float')
```

```python
median_area = bayot.Area.median()
```

```python
median_area
```

```python
bayot.Area.fillna(median_area, inplace=True)
```

# Rename columns & Translate them

```python
column_translation = {
    'نوع العقار': 'Property Type',
    'نوع العرض': 'Offer Type',
    'الرقم المرجعي': 'Reference Number',
    'حالة البناء': 'Building Status',
    'التأثيث': 'Furnishing',
    'تاريخ الإضافة': 'Date Added',
    'الملكية العقارية': 'Ownership'
}
```

```python
bayot.rename(columns=column_translation, inplace=True)
```

```python
threshold = len(bayot.columns) / 2
```

```python
bayot = bayot.dropna(thresh=threshold)
```

```python
def split_location(location):
    if pd.isna(location):  # Check if the value is NaN or missing
        return pd.Series([np.nan, np.nan, np.nan])
    
    location = str(location)
    parts = location.split('،')
    
    if len(parts) == 3:
        compound, state, city = parts[0].strip(), parts[1].strip(), parts[2].strip()
    elif len(parts) == 2: 
        compound, city = parts[0].strip(), parts[1].strip()
        state = np.nan  
    
    elif len(parts) == 1: 
        compound = parts[0].strip()
        state = city = np.nan  
    else:
        compound = state = city = np.nan  
    
    return pd.Series([compound, state, city])

bayot[['Compound Name', 'State', 'City']] = bayot['Location'].apply(split_location)

bayot = bayot.drop(columns='Location')

print(bayot)
```
